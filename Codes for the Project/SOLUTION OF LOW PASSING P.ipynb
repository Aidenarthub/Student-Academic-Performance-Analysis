#### Dealing with 'Viva' Column

```python
data['viva'].value_counts()
```

```python
data['viva'].replace(to_replace=[1,2,3,4,5],value=['Very Poor','Poor','Average','Good','Excellent'],inplace=True)
```

```python
data.viva.value_counts()
```

#### Dealing with 'Attandance' column

```python
data.columns
```

```python
# Define function to create buckets
def create_attendance_buckets(attendance):
    if attendance < 50:
        return '40-49'
    elif attendance < 60:
        return '50-59'
    elif attendance < 70:
        return '60-69'
    elif attendance < 80:
        return '70-79'
    elif attendance < 90:
        return '80-89'
    else:
        return '90-100'

# Apply function to create buckets
data['Attendance_Bucket'] = data['attandance'].apply(create_attendance_buckets)

# Display DataFrame with attendance buckets
data.head()
```

```python
data['Attendance_Bucket'].value_counts()
```
